```python
x = torch.tensor([-2.0], requires_grad=True)
y = torch.tensor([5.0], requires_grad=True)
z = torch.tensor([-4.0], requires_grad=True)

q = x + y
f = q * z

q.retain_grad()
f.retain_grad()

print(f'before backward ğŸ‘‰')
print(f'  f : {f}')
print(f'  q : {q}')
print(f'  x : {x}')
print(f'  y : {y}')
print(f'  z : {z}')

f.backward()
print(f'after  backward ğŸ‘‰')
print(f'  f : {f.grad}')
print(f'  q : {q.grad}')
print(f'  x : {x.grad}')
print(f'  y : {y.grad}')
print(f'  z : {z.grad}')
```
