````python
import torch
import numpy as np
````

### A. í…ì„œ(tensor) ì´ˆê¸°í™”
í…ì„œëŠ” ì—¬ëŸ¬ê°€ì§€ ë°©ë²•ìœ¼ë¡œ ì´ˆê¸°í™”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë‹¤ìŒ ì˜ˆë¥¼ ì‚´í´ë³´ì„¸ìš”:

#### a. ë°ì´í„°ë¡œë¶€í„° ì§ì ‘(directly) ìƒì„±í•˜ê¸°
ë°ì´í„°ë¡œë¶€í„° ì§ì ‘ í…ì„œë¥¼ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë°ì´í„°ì˜ ìë£Œí˜•(data type)ì€ ìë™ìœ¼ë¡œ ìœ ì¶”í•©ë‹ˆë‹¤.

````python
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)
x_data
````

#### b. NumPy ë°°ì—´ë¡œë¶€í„° ìƒì„±í•˜ê¸°
í…ì„œëŠ” NumPy ë°°ì—´ë¡œ ìƒì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. (ê·¸ ë°˜ëŒ€ë„ ê°€ëŠ¥í•©ë‹ˆë‹¤ - [NumPy ë³€í™˜(Bridge)](https://tutorials.pytorch.kr/beginner/blitz/tensor_tutorial.html#bridge-to-np-label) ì°¸ê³ )

````python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
x_np
````

#### c. ë‹¤ë¥¸ í…ì„œë¡œë¶€í„° ìƒì„±í•˜ê¸°
ëª…ì‹œì ìœ¼ë¡œ ì¬ì •ì˜(override)í•˜ì§€ ì•ŠëŠ”ë‹¤ë©´, ì¸ìë¡œ ì£¼ì–´ì§„ í…ì„œì˜ ì†ì„±(ëª¨ì–‘(shape), ìë£Œí˜•(datatype))ì„ ìœ ì§€í•©ë‹ˆë‹¤.

````python
x_ones = torch.ones_like(x_data) # x_dataì˜ ì†ì„±ì„ ìœ ì§€í•©ë‹ˆë‹¤.
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # x_dataì˜ ì†ì„±ì„ ë®ì–´ì”ë‹ˆë‹¤.
print(f"Random Tensor: \n {x_rand} \n")
````

#### d. ë¬´ì‘ìœ„(random) ë˜ëŠ” ìƒìˆ˜(constant) ê°’ì„ ì‚¬ìš©í•˜ê¸°
`shape`ì€ í…ì„œì˜ ì°¨ì›(dimension)ì„ ë‚˜íƒ€ë‚´ëŠ” íŠœí”Œ(tuple)ë¡œ, ì•„ë˜ í•¨ìˆ˜ë“¤ì—ì„œëŠ” ì¶œë ¥ í…ì„œì˜ ì°¨ì›ì„ ê²°ì •í•©ë‹ˆë‹¤.

````python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
````

### B. í…ì„œì˜ ì†ì„±(Attribute)
í…ì„œì˜ ì†ì„±ì€ í…ì„œì˜ ëª¨ì–‘(shape), ìë£Œí˜•(datatype) ë° ì–´ëŠ ì¥ì¹˜ì— ì €ì¥ë˜ëŠ”ì§€ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.

````python
tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
````

### C. í…ì„œ ì—°ì‚°(Operation)
ì „ì¹˜(transposing), ì¸ë±ì‹±(indexing), ìŠ¬ë¼ì´ì‹±(slicing), ìˆ˜í•™ ê³„ì‚°, ì„ í˜• ëŒ€ìˆ˜, ì„ì˜ ìƒ˜í”Œë§(random sampling) ë“±, 100ê°€ì§€ ì´ìƒì˜ í…ì„œ ì—°ì‚°ë“¤ì„ [ì—¬ê¸°](https://pytorch.org/docs/stable/torch.html)ì—ì„œ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ê° ì—°ì‚°ë“¤ì€ (ì¼ë°˜ì ìœ¼ë¡œ CPUë³´ë‹¤ ë¹ ë¥¸) GPUì—ì„œ ì‹¤í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. Colabì„ ì‚¬ìš©í•œë‹¤ë©´, Edit > Notebook Settings ì—ì„œ GPUë¥¼ í• ë‹¹í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ê¸°ë³¸ì ìœ¼ë¡œ í…ì„œëŠ” CPUì— ìƒì„±ë©ë‹ˆë‹¤. .to ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ë©´ (GPUì˜ ê°€ìš©ì„±(availability)ì„ í™•ì¸í•œ ë’¤) GPUë¡œ í…ì„œë¥¼ ëª…ì‹œì ìœ¼ë¡œ ì´ë™í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì¥ì¹˜ë“¤ ê°„ì— í° í…ì„œë“¤ì„ ë³µì‚¬í•˜ëŠ” ê²ƒì€ ì‹œê°„ê³¼ ë©”ëª¨ë¦¬ ì¸¡ë©´ì—ì„œ ë¹„ìš©ì´ ë§ì´ë“ ë‹¤ëŠ” ê²ƒì„ ê¸°ì–µí•˜ì„¸ìš”!

````python
# GPUê°€ ì¡´ì¬í•˜ë©´ í…ì„œë¥¼ ì´ë™í•©ë‹ˆë‹¤
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    print(tensor)
````

ëª©ë¡ì—ì„œ ëª‡ëª‡ ì—°ì‚°ë“¤ì„ ì‹œë„í•´ë³´ì„¸ìš”. NumPy APIì— ìµìˆ™í•˜ë‹¤ë©´ Tensor APIë¥¼ ì‚¬ìš©í•˜ëŠ” ê²ƒì€ ì‹ì€ ì£½ ë¨¹ê¸°ë¼ëŠ” ê²ƒì„ ì•Œê²Œ ë˜ì‹¤ ê²ë‹ˆë‹¤.

#### a. NumPyì‹ì˜ í‘œì¤€ ì¸ë±ì‹±ê³¼ ìŠ¬ë¼ì´ì‹±

````python
tensor = torch.ones(4, 4)
print('First row: ',tensor[0])
print('First column: ', tensor[:, 0])
print('Last column:', tensor[..., -1])
tensor[:,1] = 0
print(tensor)
````

í…ì„œ í•©ì¹˜ê¸° `torch.cat`ì„ ì‚¬ìš©í•˜ì—¬ ì£¼ì–´ì§„ ì°¨ì›ì— ë”°ë¼ ì¼ë ¨ì˜ í…ì„œë¥¼ ì—°ê²°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. torch.cat ê³¼ ë¯¸ë¬˜í•˜ê²Œ ë‹¤ë¥¸ ë˜ ë‹¤ë¥¸ í…ì„œ ê²°í•© ì—°ì‚°ì¸ [torch.stack](https://pytorch.org/docs/stable/generated/torch.stack.html)ë„ ì°¸ê³ í•´ë³´ì„¸ìš”.

````python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
````

#### b. ì‚°ìˆ  ì—°ì‚°(Arithmetic operations)

````python
# ë‘ í…ì„œ ê°„ì˜ í–‰ë ¬ ê³±(matrix multiplication)ì„ ê³„ì‚°í•©ë‹ˆë‹¤. y1, y2, y3ì€ ëª¨ë‘ ê°™ì€ ê°’ì„ ê°–ìŠµë‹ˆë‹¤.
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out=y3)


# ìš”ì†Œë³„ ê³±(element-wise product)ì„ ê³„ì‚°í•©ë‹ˆë‹¤. z1, z2, z3ëŠ” ëª¨ë‘ ê°™ì€ ê°’ì„ ê°–ìŠµë‹ˆë‹¤.
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
````

````python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
````

### E. [ì‹¤ìŠµ] Numpy to Tensor

- ì´ë•Œ, ì…ë ¥ê°’ì˜ ë²”ìœ„ëŠ” `-Ï€ ~ Ï€` ì´ë‹¤. ë”°ë¼ì„œ êµ¬í˜„ë˜ëŠ” ìˆœì „íŒŒ í•¨ìˆ˜ëŠ” ì‚¼ì°¨ë‹¤í•­ì‹ì˜ ê¼´ì´ ëœë‹¤.
- ì´ë•Œ, ì†ì‹¤ì€ ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ì˜ ìœ í´ë¦¬ë“œ ê±°ë¦¬ë¡œ êµ¬í•œë‹¤.

````python
import numpy as np
import math

# ë¬´ì‘ìœ„ë¡œ ì…ë ¥ê³¼ ì¶œë ¥ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤
x = np.linspace(-math.pi, math.pi, 2000)
y = np.sin(x)

# ë¬´ì‘ìœ„ë¡œ ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()

print(f'------------------ init ------------------')
print(f'x({len(x)}) : {x}')
print(f'y({len(y)}) : {y}')
print(f'ê°€ì¤‘ì¹˜ ì´ˆê¸°ê°’ : ')
print(f'  a : {a}')
print(f'  b : {b}')
print(f'  c : {c}')
print(f'  d : {d}')
print(f'------------------------------------------')

learning_rate = 1e-6
for t in range(2000):
    # ìˆœì „íŒŒ ë‹¨ê³„: ì˜ˆì¸¡ê°’ yë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤
    # y = a + b x + c x^2 + d x^3
    y_pred = a + b * x + c * x ** 2 + d * x ** 3

    # ì†ì‹¤(loss)ì„ ê³„ì‚°í•˜ê³  ì¶œë ¥í•©ë‹ˆë‹¤
    loss = np.square(y_pred - y).sum()
    if t % 100 == 99:
        print(f'[t = {t+1:4d}] loss : {loss:.3f}')

    # ì†ì‹¤ì— ë”°ë¥¸ a, b, c, dì˜ ë³€í™”ë„(gradient)ë¥¼ ê³„ì‚°í•˜ê³  ì—­ì „íŒŒí•©ë‹ˆë‹¤.
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()

    # ê°€ì¤‘ì¹˜ë¥¼ ê°±ì‹ í•©ë‹ˆë‹¤.
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d
    
print(f'------------------------------------------')
print(f'Result: y = {a} + {b} x + {c} x^2 + {d} x^3')
````

````python
import matplotlib.pyplot as plt

def show(x, y, y_pred):
    fig = plt.figure(figsize=(10, 4))

    plt.plot(x, y, color='red', label="real")
    plt.plot(x, y_pred, color='blue', label="prediction")
    plt.legend()
    plt.grid(True)
    plt.show()

show(x, y, a + b * x + c * x ** 2 + d * x ** 3)
````

ì´ë²ˆì—ëŠ” PyTorchì˜ ê°€ì¥ í•µì‹¬ì ì¸ ê°œë…ì¸ **í…ì„œ(Tensor)** ì— ëŒ€í•´ì„œ ì•Œì•„ë³´ê² ìŠµë‹ˆë‹¤. PyTorch í…ì„œ(Tensor)ëŠ” ê°œë…ì ìœ¼ë¡œ NumPy ë°°ì—´ê³¼ ë™ì¼í•©ë‹ˆë‹¤. 
- í…ì„œ(Tensor)ëŠ” n-ì°¨ì› ë°°ì—´ì´ë©°, PyTorchëŠ” ì´ëŸ¬í•œ í…ì„œë“¤ì˜ ì—°ì‚°ì„ ìœ„í•œ ë‹¤ì–‘í•œ ê¸°ëŠ¥ë“¤ì„ ì œê³µí•©ë‹ˆë‹¤.
- NumPy ë°°ì—´ì²˜ëŸ¼ PyTorch TensorëŠ” ë”¥ëŸ¬ë‹ì´ë‚˜ ì—°ì‚° ê·¸ë˜í”„, ë³€í™”ë„ëŠ” ì•Œì§€ ëª»í•˜ë©°, ê³¼í•™ì  ë¶„ì•¼ì˜ ì—°ì‚°ì„ ìœ„í•œ í¬ê´„ì ì¸ ë„êµ¬ì…ë‹ˆë‹¤. 
- í…ì„œëŠ” ì—°ì‚° ê·¸ë˜í”„ì™€ ë³€í™”ë„ë¥¼ ì¶”ì í•  ìˆ˜ë„ ìˆì§€ë§Œ, ê³¼í•™ì  ì—°ì‚°ì„ ìœ„í•œ ì¼ë°˜ì ì¸ ë„êµ¬ë¡œë„ ìœ ìš©í•©ë‹ˆë‹¤.
- ë˜í•œ NumPyì™€ëŠ” ë‹¤ë¥´ê²Œ, PyTorch í…ì„œëŠ” GPUë¥¼ ì‚¬ìš©í•˜ì—¬ ìˆ˜ì¹˜ ì—°ì‚°ì„ ê°€ì†í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. PyTorch í…ì„œë¥¼ GPUì—ì„œ ì‹¤í–‰í•˜ê¸° ìœ„í•´ì„œëŠ” ë‹¨ì§€ ì ì ˆí•œ ì¥ì¹˜ë¥¼ ì§€ì •í•´ì£¼ê¸°ë§Œ í•˜ë©´ ë©ë‹ˆë‹¤.

ìœ„ì˜ NumPy ì˜ˆì œì—ì„œì™€ ê°™ì´ ì‹ ê²½ë§ì˜ ìˆœì „íŒŒ ë‹¨ê³„ì™€ ì—­ì „íŒŒ ë‹¨ê³„êµ¬í˜„í•˜ë˜ ì…ë ¥/ì¶œë ¥/ê°€ì¤‘ì¹˜ ê°’ì„ PyTorch í…ì„œë¥¼ ì‚¬ìš©í•˜ì—¬ ì •ì˜í•´ ë´…ì‹œë‹¤. 

âœ… êµ¬í˜„í•´ì•¼ í•˜ëŠ” ì‚¬í•­ : ì´ì „ êµ¬í˜„ì—ì„œ Numpyë¥¼ torch.Tensorë¡œ ëŒ€ì²´í•œë‹¤.

````python
import torch
import math


dtype = torch.float
device = torch.device("cpu")
# device = torch.device("cuda:0") # GPUì—ì„œ ì‹¤í–‰í•˜ë ¤ë©´ ì´ ì£¼ì„ì„ ì œê±°í•˜ì„¸ìš”

# ë¬´ì‘ìœ„ë¡œ ì…ë ¥ê³¼ ì¶œë ¥ ë°ì´í„°ë¥¼ ìƒì„±í•©ë‹ˆë‹¤
x = torch.linspace(-math.pi, math.pi, 2000, device=device, dtype=dtype)
y = torch.sin(x)

# ë¬´ì‘ìœ„ë¡œ ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤
a = torch.randn((), device=device, dtype=dtype)
b = torch.randn((), device=device, dtype=dtype)
c = torch.randn((), device=device, dtype=dtype)
d = torch.randn((), device=device, dtype=dtype)

print(f'------------------ init ------------------')
print(f'x({len(x)}, {x.__class__}) : {x}')
print(f'y({len(y)}, {y.__class__}) : {y}')
print(f'ê°€ì¤‘ì¹˜ ì´ˆê¸°ê°’ : ')
print(f'  a({a.__class__}) : {a}')
print(f'  b({b.__class__}) : {b}')
print(f'  c({c.__class__}) : {c}')
print(f'  d({d.__class__}) : {d}')
print(f'------------------------------------------')

learning_rate = 1e-6
for t in range(2000):
    # ìˆœì „íŒŒ ë‹¨ê³„: ì˜ˆì¸¡ê°’ yë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤
    y_pred = a + b * x + c * x ** 2 + d * x ** 3

    # ì†ì‹¤(loss)ì„ ê³„ì‚°í•˜ê³  ì¶œë ¥í•©ë‹ˆë‹¤
    loss = (y_pred - y).pow(2).sum().item()
    if t % 100 == 99:
        print(f'[t = {t+1:4d}] loss : {loss:.3f}')

    # ì†ì‹¤ì— ë”°ë¥¸ a, b, c, dì˜ ë³€í™”ë„(gradient)ë¥¼ ê³„ì‚°í•˜ê³  ì—­ì „íŒŒí•©ë‹ˆë‹¤.
    grad_y_pred = 2.0 * (y_pred - y)
    grad_a = grad_y_pred.sum()
    grad_b = (grad_y_pred * x).sum()
    grad_c = (grad_y_pred * x ** 2).sum()
    grad_d = (grad_y_pred * x ** 3).sum()

    # ê°€ì¤‘ì¹˜ë¥¼ ê°±ì‹ í•©ë‹ˆë‹¤.
    a -= learning_rate * grad_a
    b -= learning_rate * grad_b
    c -= learning_rate * grad_c
    d -= learning_rate * grad_d

print(f'------------------------------------------')
print(f'Result: y = {a.item()} + {b.item()} x + {c.item()} x^2 + {d.item()} x^3')

show(x, y, a + b * x + c * x ** 2 + d * x ** 3)
````

ì‹ ê²½ë§ì„ í•™ìŠµí•  ë•Œ ê°€ì¥ ìì£¼ ì‚¬ìš©ë˜ëŠ” ì•Œê³ ë¦¬ì¦˜ì€ **ì—­ì „íŒŒ* ì…ë‹ˆë‹¤. ì´ ì•Œê³ ë¦¬ì¦˜ì—ì„œ, ë§¤ê°œë³€ìˆ˜(ëª¨ë¸ ê°€ì¤‘ì¹˜)ëŠ” ì£¼ì–´ì§„ ë§¤ê°œë³€ìˆ˜ì— ëŒ€í•œ ì†ì‹¤ í•¨ìˆ˜ì˜ **ë³€í™”ë„(gradient)**ì— ë”°ë¼ ì¡°ì •ë©ë‹ˆë‹¤.

ì´ëŸ¬í•œ ë³€í™”ë„ë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•´ PyTorchì—ëŠ” `torch.autograd`ë¼ê³  ë¶ˆë¦¬ëŠ” ìë™ ë¯¸ë¶„ ì—”ì§„ì´ ë‚´ì¥ë˜ì–´ ìˆìŠµë‹ˆë‹¤. ì´ëŠ” ëª¨ë“  ê³„ì‚° ê·¸ë˜í”„ì— ëŒ€í•œ ë³€í™”ë„ì˜ ìë™ ê³„ì‚°ì„ ì§€ì›í•©ë‹ˆë‹¤.

ì…ë ¥ `x`, ë§¤ê°œë³€ìˆ˜ `w` ì™€ `b` , ê·¸ë¦¬ê³  ì¼ë¶€ ì†ì‹¤ í•¨ìˆ˜ê°€ ìˆëŠ” ê°€ì¥ ê°„ë‹¨í•œ ë‹¨ì¼ ê³„ì¸µ ì‹ ê²½ë§ì„ ê°€ì •í•˜ê² ìŠµë‹ˆë‹¤. PyTorchì—ì„œëŠ” ë‹¤ìŒê³¼ ê°™ì´ ì •ì˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
import torch

x = torch.ones(5)  # input tensor
y = torch.zeros(3)  # expected output
w = torch.randn(5, 3, requires_grad=True)
b = torch.randn(3, requires_grad=True)
z = torch.matmul(x, w)+b
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)
````

### C. ë³€í™”ë„ ì¶”ì  ë©ˆì¶”ê¸°

- ë³€í™”ë„ë¥¼ ì¶”ì í•˜ì§€ ì•ŠëŠ” í…ì„œì˜ ì—°ì‚°ì´ ë” íš¨ìœ¨ì ì´ê¸° ë•Œë¬¸ì—, ìˆœì „íŒŒ ë‹¨ê³„ë§Œ ìˆ˜í–‰í•  ë•Œ **ì—°ì‚° ì†ë„ê°€ í–¥ìƒë©ë‹ˆë‹¤.**

### D. ì—°ì‚° ê·¸ë˜í”„ì— ëŒ€í•œ ì¶”ê°€ ì •ë³´

ê°œë…ì ìœ¼ë¡œ, autogradëŠ” ë°ì´í„°(í…ì„œ)ì˜ ë° ì‹¤í–‰ëœ ëª¨ë“  ì—°ì‚°ë“¤(ë° ì—°ì‚° ê²°ê³¼ê°€ ìƒˆë¡œìš´ í…ì„œì¸ ê²½ìš°ë„ í¬í•¨í•˜ì—¬)ì˜ ê¸°ë¡ì„ [Function](https://pytorch.org/docs/stable/autograd.html#torch.autograd.Function) ê°ì²´ë¡œ
êµ¬ì„±ëœ ë°©í–¥ì„± ë¹„ìˆœí™˜ ê·¸ë˜í”„(DAG; Directed Acyclic Graph)ì— ì €ì¥(keep)í•©ë‹ˆë‹¤. ì´ ë°©í–¥ì„± ë¹„ìˆœí™˜ ê·¸ë˜í”„(DAG)ì˜ ì(leave)ì€ ì…ë ¥ í…ì„œì´ê³ , ë¿Œë¦¬(root)ëŠ” ê²°ê³¼ í…ì„œì…ë‹ˆë‹¤. ì´ ê·¸ë˜í”„ë¥¼ ë¿Œë¦¬ì—ì„œë¶€í„° ìê¹Œì§€ ì¶”ì í•˜ë©´ ì—°ì‡„ ë²•ì¹™(chain rule)ì— ë”°ë¼ ë³€í™”ë„ë¥¼ ìë™ìœ¼ë¡œ ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ìˆœì „íŒŒ ë‹¨ê³„ì—ì„œ, `autograd`ëŠ” ë‹¤ìŒ ë‘ ê°€ì§€ ì‘ì—…ì„ ë™ì‹œì— ìˆ˜í–‰í•©ë‹ˆë‹¤:

- ìš”ì²­ëœ ì—°ì‚°ì„ ìˆ˜í–‰í•˜ì—¬ ê²°ê³¼ í…ì„œë¥¼ ê³„ì‚°í•˜ê³ ,
- DAGì— ì—°ì‚°ì˜ *ë³€í™”ë„ ê¸°ëŠ¥(gradient function)* ë¥¼ ìœ ì§€(maintain)í•©ë‹ˆë‹¤.

ì—­ì „íŒŒ ë‹¨ê³„ëŠ” DAG ë¿Œë¦¬(root)ì—ì„œ ``.backward()`` ê°€ í˜¸ì¶œë  ë•Œ ì‹œì‘ë©ë‹ˆë‹¤. ``autograd`` ëŠ” ì´ ë•Œ:

- ê° ``.grad_fn`` ìœ¼ë¡œë¶€í„° ë³€í™”ë„ë¥¼ ê³„ì‚°í•˜ê³ ,
- ê° í…ì„œì˜ ``.grad`` ì†ì„±ì— ê³„ì‚° ê²°ê³¼ë¥¼ ìŒ“ê³ (accumulate),
- ì—°ì‡„ ë²•ì¹™ì„ ì‚¬ìš©í•˜ì—¬, ëª¨ë“  ì(leaf) í…ì„œë“¤ê¹Œì§€ ì „íŒŒ(propagate)í•©ë‹ˆë‹¤.

### E. ì„ íƒì ìœ¼ë¡œ ì½ê¸°(Optional Reading): í…ì„œ ë³€í™”ë„ì™€ ì•¼ì½”ë¹„ì•ˆ ê³±(Jacobian Product)

ëŒ€ë¶€ë¶„ì˜ ê²½ìš°, ìŠ¤ì¹¼ë¼ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ê°€ì§€ê³  ì¼ë¶€ ë§¤ê°œë³€ìˆ˜ì™€ ê´€ë ¨í•œ ë³€í™”ë„ë¥¼ ê³„ì‚°í•´ì•¼ í•©ë‹ˆë‹¤. ê·¸ëŸ¬ë‚˜ ì¶œë ¥ í•¨ìˆ˜ê°€ ì„ì˜ì˜ í…ì„œì¸ ê²½ìš°ê°€ ìˆìŠµë‹ˆë‹¤. ì´ëŸ´ ë•Œ, PyTorchëŠ” ì‹¤ì œ ë³€í™”ë„ê°€ ì•„ë‹Œ **ì•¼ì½”ë¹„ì•ˆ ê³±(Jacobian product)**\ ì„ ê³„ì‚°í•©ë‹ˆë‹¤.

$\vec{x}=\langle x_1,\dots,x_n\rangle$ ì´ê³ , $\vec{y}=\langle y_1,\dots,y_m\rangle`$ ì¼ ë•Œ ë²¡í„° í•¨ìˆ˜ $\vec{y}=f(\vec{x})$ì—ì„œ $\vec{x}$ì— ëŒ€í•œ $\vec{y}$ì˜ ë³€í™”ë„ëŠ” **ì•¼ì½”ë¹„ì•ˆ í–‰ë ¬(Jacobian matrix)**ë¡œ ì£¼ì–´ì§‘ë‹ˆë‹¤:

$$J=\left(\begin{array}{ccc}
      \frac{\partial y_{1}}{\partial x_{1}} & \cdots & \frac{\partial y_{1}}{\partial x_{n}}\\
      \vdots & \ddots & \vdots\\
      \frac{\partial y_{m}}{\partial x_{1}} & \cdots & \frac{\partial y_{m}}{\partial x_{n}}
      \end{array}\right) $$

ì•¼ì½”ë¹„ì•ˆ í–‰ë ¬ ìì²´ë¥¼ ê³„ì‚°í•˜ëŠ” ëŒ€ì‹ , PyTorchëŠ” ì£¼ì–´ì§„ ì…ë ¥ ë²¡í„° $v=(v_1 \dots v_m)$ ì— ëŒ€í•œ **ì•¼ì½”ë¹„ì•ˆ ê³±(Jacobian Product)**  $v^T\cdot J$ ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
ì´ ê³¼ì •ì€ $v$ ë¥¼ ì¸ìë¡œ ``backward`` ë¥¼ í˜¸ì¶œí•˜ë©´ ì´ë¤„ì§‘ë‹ˆë‹¤. $v$ì˜ í¬ê¸°ëŠ” ê³±(product)ì„ ê³„ì‚°í•˜ë ¤ê³  í•˜ëŠ” ì›ë˜ í…ì„œì˜ í¬ê¸°ì™€ ê°™ì•„ì•¼ í•©ë‹ˆë‹¤.

````python
inp = torch.eye(5, requires_grad=True)
out = (inp+1).pow(2)
out.backward(torch.ones_like(inp), retain_graph=True)
print("First call\n", inp.grad)
out.backward(torch.ones_like(inp), retain_graph=True)
print("\nSecond call\n", inp.grad)
inp.grad.zero_()
out.backward(torch.ones_like(inp), retain_graph=True)
print("\nCall after zeroing gradients\n", inp.grad)
````

ë™ì¼í•œ ì¸ìë¡œ `backward`ë¥¼ ë‘ì°¨ë¡€ í˜¸ì¶œí•˜ë©´ ë³€í™”ë„ ê°’ì´ ë‹¬ë¼ì§‘ë‹ˆë‹¤. ì´ëŠ” ì—­ë°©í–¥ ì „íŒŒë¥¼ ìˆ˜í–‰í•  ë•Œ, PyTorchê°€ **ë³€í™”ë„ë¥¼ ëˆ„ì (accumulate)í•´ë‘ê¸° ë•Œë¬¸**ì…ë‹ˆë‹¤. ì¦‰, ê³„ì‚°ëœ ë³€í™”ë„ì˜ ê°’ì´ ì—°ì‚° ê·¸ë˜í”„ì˜ ëª¨ë“  ì(leaf) ë…¸ë“œì˜ `grad` ì†ì„±ì— ì¶”ê°€ë©ë‹ˆë‹¤. ë”°ë¼ì„œ ì œëŒ€ë¡œëœ ë³€í™”ë„ë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•´ì„œëŠ” `grad` ì†ì„±ì„ ë¨¼ì € 0ìœ¼ë¡œ ë§Œë“¤ì–´ì•¼ í•©ë‹ˆë‹¤. ì‹¤ì œ í•™ìŠµ ê³¼ì •ì—ì„œëŠ” *ì˜µí‹°ë§ˆì´ì €(optimizer)* ê°€ ì´ ê³¼ì •ì„ ë„ì™€ì¤ë‹ˆë‹¤.

### F. [ì‹¤ìŠµ] Backpropagation with Autograd

ì•ì„  ì‹¤ìŠµì—ì„œëŠ” ì‹ ê²½ë§ì˜ ìˆœì „íŒŒ ë‹¨ê³„ì™€ ì—­ì „íŒŒ ë‹¨ê³„ë¥¼ ì§ì ‘ êµ¬í˜„í•´ë³´ì•˜ìŠµë‹ˆë‹¤. ì‘ì€ 2ê³„ì¸µ(2-layer) ì‹ ê²½ë§ì—ì„œëŠ” ì—­ì „íŒŒ ë‹¨ê³„ë¥¼ ì§ì ‘ êµ¬í˜„í•˜ëŠ” ê²ƒì´ í°ì¼ì´ ì•„ë‹ˆì§€ë§Œ, ë³µì¡í•œ ëŒ€ê·œëª¨ ì‹ ê²½ë§ì—ì„œëŠ” ë§¤ìš° ì•„ìŠ¬ì•„ìŠ¬í•œ ì¼ì¼ ê²ƒì…ë‹ˆë‹¤.

ë‹¤í–‰íˆë„, [ìë™ ë¯¸ë¶„](https://en.wikipedia.org/wiki/Automatic_differentiation) ì„ ì‚¬ìš©í•˜ì—¬ ì‹ ê²½ë§ì˜ ì—­ì „íŒŒ ë‹¨ê³„ ì—°ì‚°ì„ ìë™í™”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. PyTorchì˜ **autograd** íŒ¨í‚¤ì§€ëŠ” ì •í™•íˆ ì´ëŸ° ê¸°ëŠ¥ì„ ì œê³µí•©ë‹ˆë‹¤. Autogradë¥¼ ì‚¬ìš©í•˜ë©´, ì‹ ê²½ë§ì˜ ìˆœì „íŒŒ ë‹¨ê³„ì—ì„œ **ì—°ì‚° ê·¸ë˜í”„(computational graph)** ë¥¼ ì •ì˜í•˜ê²Œ ë©ë‹ˆë‹¤; ì´ ê·¸ë˜í”„ì˜ ë…¸ë“œ(node)ëŠ” í…ì„œ(tensor)ì´ê³ , ì—£ì§€(edge)ëŠ” ì…ë ¥ í…ì„œë¡œë¶€í„° ì¶œë ¥ í…ì„œë¥¼ ë§Œë“¤ì–´ë‚´ëŠ” í•¨ìˆ˜ê°€ ë©ë‹ˆë‹¤. ì´ ê·¸ë˜í”„ë¥¼ í†µí•´ ì—­ì „íŒŒë¥¼ í•˜ê²Œ ë˜ë©´ ë³€í™”ë„ë¥¼ ì‰½ê²Œ ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ì´ëŠ” ë³µì¡í•˜ê²Œ ë“¤ë¦¬ê² ì§€ë§Œ, ì‹¤ì œë¡œ ì‚¬ìš©í•˜ëŠ” ê²ƒì€ ë§¤ìš° ê°„ë‹¨í•©ë‹ˆë‹¤. ê° í…ì„œëŠ” ì—°ì‚°ê·¸ë˜í”„ì—ì„œ ë…¸ë“œë¡œ í‘œí˜„ë©ë‹ˆë‹¤. ë§Œì•½ x ê°€ `x.requires_grad=True` ì¸ í…ì„œë¼ë©´ `x.grad` ì–´ë–¤ ìŠ¤ì¹¼ë¼ ê°’ì— ëŒ€í•œ x ì˜ ë³€í™”ë„ë¥¼ ê°–ëŠ” ë˜ ë‹¤ë¥¸ í…ì„œì…ë‹ˆë‹¤.

ì´ë²ˆ ë‹¨ê³„ì—ì„œëŠ” í…ì„œ ì—°ì‚°ì„ ì‚¬ìš©í•˜ì—¬ ìˆœì „íŒŒ ë‹¨ê³„ë¥¼ ê³„ì‚°í•˜ê³ , PyTorch autogradë¥¼ ì‚¬ìš©í•˜ì—¬ ë³€í™”ë„(gradient)ë¥¼ ê³„ì‚°í•´ë³´ê² ìŠµë‹ˆë‹¤.

âœ… êµ¬í˜„í•´ì•¼ í•˜ëŠ” ì‚¬í•­ : ì´ì „ êµ¬í˜„ì—ì„œ ì—­ì „íŒŒ ë¶€ë¶„ì„ autograd ê¸°ëŠ¥ìœ¼ë¡œ ëŒ€ì²´í•œë‹¤.

````python
import torch
import math

dtype = torch.float
device = torch.device("cpu")
# device = torch.device("cuda:0") # GPUì—ì„œ ì‹¤í–‰í•˜ë ¤ë©´ ì´ ì£¼ì„ì„ ì œê±°í•˜ì„¸ìš”

# ì…ë ¥ê°’ê³¼ ì¶œë ¥ê°’ì„ ê°–ëŠ” í…ì„œë“¤ì„ ìƒì„±í•©ë‹ˆë‹¤.
# requires_grad=Falseê°€ ê¸°ë³¸ê°’ìœ¼ë¡œ ì„¤ì •ë˜ì–´ ì—­ì „íŒŒ ë‹¨ê³„ ì¤‘ì— ì´ í…ì„œë“¤ì— ëŒ€í•œ ë³€í™”ë„ë¥¼
# ê³„ì‚°í•  í•„ìš”ê°€ ì—†ìŒì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
x = torch.linspace(-math.pi, math.pi, 2000, device=device, dtype=dtype)
y = torch.sin(x)

# ê°€ì¤‘ì¹˜ë¥¼ ê°–ëŠ” ì„ì˜ì˜ í…ì„œë¥¼ ìƒì„±í•©ë‹ˆë‹¤. 3ì°¨ ë‹¤í•­ì‹ì´ë¯€ë¡œ 4ê°œì˜ ê°€ì¤‘ì¹˜ê°€ í•„ìš”í•©ë‹ˆë‹¤:
# y = a + b x + c x^2 + d x^3
# requires_grad=Trueë¡œ ì„¤ì •í•˜ì—¬ ì—­ì „íŒŒ ë‹¨ê³„ ì¤‘ì— ì´ í…ì„œë“¤ì— ëŒ€í•œ ë³€í™”ë„ë¥¼ ê³„ì‚°í•  í•„ìš”ê°€ 
# ìˆìŒì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
a = torch.randn((), device=device, dtype=dtype, requires_grad=True)
b = torch.randn((), device=device, dtype=dtype, requires_grad=True)
c = torch.randn((), device=device, dtype=dtype, requires_grad=True)
d = torch.randn((), device=device, dtype=dtype, requires_grad=True)

print(f'------------------ init ------------------')
print(f'x({len(x)}) : {x}')
print(f'y({len(y)}) : {y}')
print(f'ê°€ì¤‘ì¹˜ ì´ˆê¸°ê°’ : ')
print(f'  a : {a}')
print(f'  b : {b}')
print(f'  c : {c}')
print(f'  d : {d}')
print(f'------------------------------------------')

learning_rate = 1e-6
for t in range(2000):
    # ìˆœì „íŒŒ ë‹¨ê³„: í…ì„œë“¤ ê°„ì˜ ì—°ì‚°ì„ ì‚¬ìš©í•˜ì—¬ ì˜ˆì¸¡ê°’ yë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    y_pred = a + b * x + c * x ** 2 + d * x ** 3

    # í…ì„œë“¤ê°„ì˜ ì—°ì‚°ì„ ì‚¬ìš©í•˜ì—¬ ì†ì‹¤(loss)ì„ ê³„ì‹¼í•˜ê³  ì¶œë ¥í•©ë‹ˆë‹¤.
    # ì´ ë•Œ ì†ì‹¤ì€ (1,) shapeì„ ê°–ëŠ” í…ì„œì…ë‹ˆë‹¤.
    # loss.item() ìœ¼ë¡œ ì†ì‹¤ì´ ê°–ê³  ìˆëŠ” ìŠ¤ì¹¼ë¼ ê°’ì„ ê°€ì ¸ì˜¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
    loss = (y_pred - y).pow(2).sum()
    if t % 100 == 99:
        print(f'[t = {t+1:4d}] loss : {loss:.3f}')

    # autograd ë¥¼ ì‚¬ìš©í•˜ì—¬ ì—­ì „íŒŒ ë‹¨ê³„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. ì´ëŠ” requires_grad=Trueë¥¼ ê°–ëŠ”
    # ëª¨ë“  í…ì„œë“¤ì— ëŒ€í•œ ì†ì‹¤ì˜ ë³€í™”ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. 
    # ì´í›„ a.gradì™€ b.grad, c.grad, d.gradëŠ” ê°ê° a, b, c, dì— ëŒ€í•œ ì†ì‹¤ì˜ ë³€í™”ë„ë¥¼
    # ê°–ëŠ” í…ì„œê°€ ë©ë‹ˆë‹¤.
    loss.backward()

    # ê²½ì‚¬í•˜ê°•ë²•(gradient descent)ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°€ì¤‘ì¹˜ë¥¼ ì§ì ‘ ê°±ì‹ í•©ë‹ˆë‹¤.
    # torch.no_grad()ë¡œ ê°ì‹¸ëŠ” ì´ìœ ëŠ”, ê°€ì¤‘ì¹˜ë“¤ì´ requires_grad=True ì§€ë§Œ
    # autogradì—ì„œëŠ” ì´ë¥¼ ì¶”ì í•˜ì§€ ì•Šì„ ê²ƒì´ê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.
    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad
        c -= learning_rate * c.grad
        d -= learning_rate * d.grad

        # ê°€ì¤‘ì¹˜ ê°±ì‹  í›„ì—ëŠ” ë³€í™”ë„ë¥¼ ì§ì ‘ 0ìœ¼ë¡œ ë§Œë“­ë‹ˆë‹¤.
        a.grad = None
        b.grad = None
        c.grad = None
        d.grad = None

print(f'------------------------------------------')
print(f'Result: y = {a.item()} + {b.item()} x + {c.item()} x^2 + {d.item()} x^3')

with torch.no_grad():
    show(x, y, a + b * x + c * x ** 2 + d * x ** 3)
````

## 3. DATASETê³¼ DATALOADER
---

ë°ì´í„° ìƒ˜í”Œì„ ì²˜ë¦¬í•˜ëŠ” ì½”ë“œëŠ” ì§€ì €ë¶„(messy)í•˜ê³  ìœ ì§€ë³´ìˆ˜ê°€ ì–´ë ¤ìš¸ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë” ë‚˜ì€ ê°€ë…ì„±(readability)ê³¼ ëª¨ë“ˆì„±(modularity)ì„ ìœ„í•´ ë°ì´í„°ì…‹ ì½”ë“œë¥¼ ëª¨ë¸ í•™ìŠµ ì½”ë“œë¡œë¶€í„° ë¶„ë¦¬í•˜ëŠ” ê²ƒì´ ì´ìƒì ì…ë‹ˆë‹¤. ë‹¤ìŒ ë§¤ê°œë³€ìˆ˜ë“¤ì„ ì‚¬ìš©í•˜ì—¬ [FashionMNIST ë°ì´í„°ì…‹](https://pytorch.org/vision/stable/datasets.html#fashion-mnist)ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤:
 - ``root`` ëŠ” í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë°ì´í„°ê°€ ì €ì¥ë˜ëŠ” ê²½ë¡œì…ë‹ˆë‹¤.
 - ``train`` ì€ í•™ìŠµìš© ë˜ëŠ” í…ŒìŠ¤íŠ¸ìš© ë°ì´í„°ì…‹ ì—¬ë¶€ë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
 - ``download=True`` ëŠ” ``root`` ì— ë°ì´í„°ê°€ ì—†ëŠ” ê²½ìš° ì¸í„°ë„·ì—ì„œ ë‹¤ìš´ë¡œë“œí•©ë‹ˆë‹¤.
 - ``transform`` ê³¼ ``target_transform`` ì€ íŠ¹ì§•(feature)ê³¼ ì •ë‹µ(label) ë³€í˜•(transform)ì„ ì§€ì •í•©ë‹ˆë‹¤.

````python
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt


training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)
````

### C. íŒŒì¼ì—ì„œ ì‚¬ìš©ì ì •ì˜ ë°ì´í„°ì…‹ ë§Œë“¤ê¸°

ì‚¬ìš©ì ì •ì˜ Dataset í´ë˜ìŠ¤ëŠ” ë°˜ë“œì‹œ 3ê°œ í•¨ìˆ˜ë¥¼ êµ¬í˜„í•´ì•¼ í•©ë‹ˆë‹¤: `__init__`, `__len__`, and `__getitem__`. ì•„ë˜ êµ¬í˜„ì„ ì‚´í´ë³´ë©´ FashionMNIST ì´ë¯¸ì§€ë“¤ì€ img_dir ë””ë ‰í† ë¦¬ì— ì €ì¥ë˜ê³ , ì •ë‹µì€ `annotations_file csv` íŒŒì¼ì— ë³„ë„ë¡œ ì €ì¥ë©ë‹ˆë‹¤.

ë‹¤ìŒ ì¥ì—ì„œ ê° í•¨ìˆ˜ë“¤ì—ì„œ ì¼ì–´ë‚˜ëŠ” ì¼ë“¤ì„ ìì„¸íˆ ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤

````python
import os
import pandas as pd
from torchvision.io import read_image

class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file)
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
````

#### a. \_\_init\_\_

`__init__` í•¨ìˆ˜ëŠ” Dataset ê°ì²´ê°€ ìƒì„±(instantiate)ë  ë•Œ í•œ ë²ˆë§Œ ì‹¤í–‰ë©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ì´ë¯¸ì§€ì™€ ì£¼ì„ íŒŒì¼(annotation_file)ì´ í¬í•¨ëœ ë””ë ‰í† ë¦¬ì™€ (ë‹¤ìŒ ì¥ì—ì„œ ìì„¸íˆ ì‚´í´ë³¼) ë‘ê°€ì§€ ë³€í˜•(transform)ì„ ì´ˆê¸°í™”í•©ë‹ˆë‹¤. 

labels.csv íŒŒì¼ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤: 
```
tshirt1.jpg, 0

......

```

````python
def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
    self.img_labels = pd.read_csv(annotations_file)
    self.img_dir = img_dir
    self.transform = transform
    self.target_transform = target_transform
````

#### b. \_\_len\_\_

`__len__` í•¨ìˆ˜ëŠ” ë°ì´í„°ì…‹ì˜ ìƒ˜í”Œ ê°œìˆ˜ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.

ì˜ˆ:

````python
def __len__(self):
    return len(self.img_labels)
````

#### c. \_\_getitem\_\_

`__getitem__` í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ì¸ë±ìŠ¤ ``idx`` ì— í•´ë‹¹í•˜ëŠ” ìƒ˜í”Œì„ ë°ì´í„°ì…‹ì—ì„œ ë¶ˆëŸ¬ì˜¤ê³  ë°˜í™˜í•©ë‹ˆë‹¤.<br>
ì¸ë±ìŠ¤ë¥¼ ê¸°ë°˜ìœ¼ë¡œ, ë””ìŠ¤í¬ì—ì„œ ì´ë¯¸ì§€ì˜ ìœ„ì¹˜ë¥¼ ì‹ë³„í•˜ê³ , ``read_image`` ë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ë¯¸ì§€ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³ , ``self.img_labels`` ì˜ csv ë°ì´í„°ë¡œë¶€í„° í•´ë‹¹í•˜ëŠ” ì •ë‹µ(label)ì„ ê°€ì ¸ì˜¤ê³ , (í•´ë‹¹í•˜ëŠ” ê²½ìš°) ë³€í˜•(transform) í•¨ìˆ˜ë“¤ì„ í˜¸ì¶œí•œ ë’¤, í…ì„œ ì´ë¯¸ì§€ì™€ ë¼ë²¨ì„ Python ì‚¬ì „(dict)í˜•ìœ¼ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.

````python
def __getitem__(self, idx):
    img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
    image = read_image(img_path)
    label = self.img_labels.iloc[idx, 1]
    if self.transform:
        image = self.transform(image)
    if self.target_transform:
        label = self.target_transform(label)
    sample = {"image": image, "label": label}
    return sample
````

### D. DataLoaderë¡œ í•™ìŠµìš© ë°ì´í„° ì¤€ë¹„í•˜ê¸°

`Dataset`ì€ ë°ì´í„°ì…‹ì˜ íŠ¹ì§•(feature)ì„ ê°€ì ¸ì˜¤ê³  í•˜ë‚˜ì˜ ìƒ˜í”Œì— ì •ë‹µ(label)ì„ ì§€ì •í•˜ëŠ” ì¼ì„ í•œ ë²ˆì— í•©ë‹ˆë‹¤. ëª¨ë¸ì„ í•™ìŠµí•  ë•Œ, ì¼ë°˜ì ìœ¼ë¡œ ìƒ˜í”Œë“¤ì„ â€œë¯¸ë‹ˆë°°ì¹˜(minibatch)â€ë¡œ ì „ë‹¬í•˜ê³ , ë§¤ ì—í­(epoch)ë§ˆë‹¤ ë°ì´í„°ë¥¼ ë‹¤ì‹œ ì„ì–´ì„œ ê³¼ì í•©(overfit)ì„ ë§‰ê³ , Pythonì˜ multiprocessing ì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„° ê²€ìƒ‰ ì†ë„ë¥¼ ë†’ì´ë ¤ê³  í•©ë‹ˆë‹¤.

`DataLoader`ëŠ” ê°„ë‹¨í•œ APIë¡œ ì´ëŸ¬í•œ ë³µì¡í•œ ê³¼ì •ë“¤ì„ ì¶”ìƒí™”í•œ ìˆœíšŒ ê°€ëŠ¥í•œ ê°ì²´(iteratable)ì…ë‹ˆë‹¤.

````python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(training_data, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=64, shuffle=True)

train_dataloader, test_dataloader
````

### E. DataLoaderë¥¼ í†µí•´ ìˆœíšŒí•˜ê¸°(iterate)

````python
# ì´ë¯¸ì§€ì™€ ì •ë‹µ(label)ì„ í‘œì‹œí•©ë‹ˆë‹¤.
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0].squeeze()
label = train_labels[0]
plt.imshow(img, cmap="gray")
plt.show()
print(f"Label: {label}")
````
